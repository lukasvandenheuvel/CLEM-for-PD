```python
# For Amanda Amsterdam images
import numpy as np
from skimage import io
img = reader.lif.get_image(f)
for ch in range(img.channels):
    for z in range(img.dims.z):
        c = (ch*img.dims.z+z) % img.channels
        real_z = (ch*img.dims.z+z) // img.channels

        # Make folder to save tiles
        channel_path = os.path.join(output_path, f'ch{c}_s0_zstack')
        if not (os.path.exists(channel_path)):
            os.mkdir(channel_path)
            print('Made new channels folder')
        else:
            print('Overwriting existing tiles')

        img_name = f'ch{int(c)}_s0_z{int(real_z)}.tif'
        frame = np.array(img.get_frame(z=z,t=0,c=ch,m=0))
        io.imsave(os.path.join(channel_path,f'{img_name}'), frame)
        print(f'Saved img {img_name}')
```
